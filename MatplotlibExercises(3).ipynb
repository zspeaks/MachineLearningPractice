# Matplotlib exercises

## Exercise 1

Plot the functions $\sqrt{x}$, $x$ and $x^2$ in the interval $x\in (0,1)$ on the same plot. Use different line styles and include an appropriate legend, plot title and labels. Use equal aspect ratio on the x and y axes.

```
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,1,100)
plt.plot(x, np.sqrt(x),"-", label = "x^(1/2)")
plt.plot(x, x, "--", label = "x")
plt.plot(x, x**2, "-.", label = "x^2")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Comparison of rates of growth")
plt.axes().set_aspect('equal')


plt.show()
```

## Exercise 2

Make a scatter plot showing two populations of 100 points each. Both populations are sampled from a two-dimensional Gaussian. For the first population, let the Gaussian be centered at (0,0) and have $\sigma_x=\sigma_y=1$. For the second population, let the Gaussian be centered at (2,2) and have $\sigma_x=\sigma_y=3$.

```
x1 = np.random.normal(1, 1, 100)
y1 = np.random.normal(1, 1, 100)
x2 = np.random.normal(2, 3, 100)
y2 = np.random.normal(2, 3, 100)

plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.show()
```

## Exercise 3

Create two histograms of the $x$-values and the $y$-values of the two populations in the previous exercise. 

```
plt.hist([x1,x2])
plt.title("X")
plt.show()
plt.hist([y1,y2])
plt.title("Y")
plt.show()
```
